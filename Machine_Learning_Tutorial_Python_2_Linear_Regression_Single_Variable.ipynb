```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#sklearn is a library for ML models
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/drive/MyDrive/008-Interviews/Machine_Learning_Projects/Data/homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('price (US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
#training
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=m*x+b
135.78767123*3300+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
#exercise
#Predict canada's per capita income in year 2020. There is an exercise
#folder here on github at same level as this notebook, download that
#and you will find canada_per_capita_income.csv file. Using this build
#a regression model and predict the per capita income for canadian citizens in year 2020
df = pd.read_csv('/content/drive/MyDrive/008-Interviews/Machine_Learning_Projects/Data/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
#training
reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year', fontsize=20)
plt.ylabel('per capita income', fontsize=20)
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```

```python
reg.coef_
```

```python
reg.intercept_
```
